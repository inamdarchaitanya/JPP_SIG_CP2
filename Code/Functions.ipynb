```python
#Function to compare 2 numbers
def Compare(a, b):
    if a > b:
        return f"{a} is greater than {b}"
    elif b > a:
        return f"{b} is greater than {a}"
    else:
        return "Both numbers are equal"

#Function to generate tables in a given range
def GenerateTable(num1,num2):
    for i in range(num1, num2+1):
        print(f"Table of {i}:")
        for j in range(1, 11):
            print(f"{i} x {j} = {i * j}")
        print()  # Blank line for separation
        
#Function to convert celcius to Farenheit
def Convert(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return f"{celsius}째C is equal to {fahrenheit}째F"

#Function to calculate area of a rectangle
def CalculateArea(length, width):
    area = length * width
    return f"The area of the rectangle is {area} square units"

#Function to check if a given number is prime 
def CheckifPrime(number):
    if number <= 1:
        return f"{number} is not a prime number"
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return f"{number} is not a prime number"
    return f"{number} is a prime number"


print(Compare(18,9))
GenerateTable(2,10)
print(Convert(100))
print(CalculateArea(10,20))
print (CheckifPrime(18))
print (CheckifPrime(97))
```
